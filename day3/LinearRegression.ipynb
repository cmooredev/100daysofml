```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate synthetic data
np.random.seed(0)
#create an array with shape (100,1) filled with random numbers 2*(0-1)
x = 2 * np.random.rand(100,1)
#create array with shape (100,1) filled from standard normal distribution 
y = 4 + 3 * x + np.random.randn(100,1)

#visualize
plt.scatter(x,y)
plt.title('Scatter x vs y')
plt.show()
```

```python
#predict returns the y value for x given b_0, b_1
def predict (x, b_0, b_1):
    return b_0 + b_1 * x
```

```python
#compute the error with MSE by 
def compute_cost(x, y, b_0, b_1):
    m = len(y)
    #get predicted y values for given x and current b_0, b_1
    y_pred = predict(x, b_0, b_1)
    #find the mean squared error
    cost = (1/(2*m)) * np.sum((y_pred - y)**2)
    return cost
```

```python
def gradient_descent(x, y, b_0, b_1, learning_rate, n_iterations):
    m = len(y)
    #load with zeros
    cost_history = np.zeros(n_iterations)

    #loop for n iterations
    for i in range(n_iterations):
        #find prediction
        y_pred = predict(x, b_0, b_1)
        #adjust based on learning rate and current error
        b_0 -= learning_rate * (1/m) * np.sum(y_pred - y)
        b_1 -= learning_rate * (1/m) * np.sum((y_pred - y) * x)
        #record cost history
        cost_history[i] = compute_cost(x, y, b_0, b_1)

    return b_0, b_1, cost_history
```

```python
#set intercept and slope to random values
b_0 = np.random.randn(1)
b_1 = np.random.randn(1)
#set learning rate and iterations
learning_rate = 0.1
n_iterations = 100

#compute values
b_0, b_1, cost_history = gradient_descent(x, y, b_0, b_1, learning_rate, n_iterations)
```

```python
plt.plot(range(n_iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost function over iterations')
plt.show()
```

```python
plt.scatter(x, y, label='Data')
plt.plot(x, predict(x, b_0, b_1), color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
#try changing learning rate and iterations to see how the cost function changes over iterations and the final regression fit
learning_rate = 0.05
n_iterations = 15

#reset b_0, b_1
b_0 = np.random.randn(1)
b_1 = np.random.randn(1)

#compute values
b_0, b_1, cost_history = gradient_descent(x, y, b_0, b_1, learning_rate, n_iterations)

plt.plot(range(n_iterations), cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost function over iterations')
plt.show()

plt.scatter(x, y, label='Data')
plt.plot(x, predict(x, b_0, b_1), color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
